```python
import pandas as pd
import numpy as np
```

```python
data = {'國家':[],
        '人口':[]}
data = pd.DataFrame(data)
```

```python
Country = ['天使','天使宮','經濟特區·日本','塔利比亞共和國','MSWAD',
           '鋼彈調查隊','艾莉絲公司','超兵機關','頂武 ','摩拉利亞共和國',
           '聯合國軍','地球聯邦','太陽勇者隊','A-Laws','純源',
           '變革者','阿札迪斯坦王國','庫爾吉斯共和國','蘇伊爾王國','里切拉王國',
           '真愛爾蘭共和軍','PMC Trust','拉·伊甸拉','KPSA']
```

```python
import random
all_num = 1000000
a_list = [i for i in range(all_num)]
num = len(Country)

def make_random():
    b_length = 1
    b_list = []
    while b_length<=num:
        random_number = random.randint(1, all_num - 1)
        if random_number not in b_list:
            b_list.append(random_number)
            b_length = len(b_list)+1
    return b_list
population = make_random()

zipped = list(zip(Country,population))
for i in range(len(zipped)):
    data.loc[i] = zipped[i]
    
max_Country = data['國家'][data['人口'].idxmax()]
print('最大的國家為 %s' % max_Country + ' 人口數量為 %s' % data['人口'].max() )
data
```
